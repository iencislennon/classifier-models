```python
import pandas as pd 
df = pd.read_csv('/Users/ansartleubayev/Documents/edas/classification_dataset.csv')
df.head()
```

```python
X = df.drop(columns=['Target'])
y = df.Target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# knn
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=8, algorithm='ball_tree')
knn.fit(X_train, y_train)
y_pread1 = knn.predict(X_test)
#log reg
from sklearn.linear_model import LogisticRegression
lg = LogisticRegression(max_iter=100)
lg.fit(X_train, y_train)
y_pread2 = lg.predict(X_test)
#svm
from sklearn.svm import SVC 
svc = SVC( kernel='linear', C=1.0)
svc.fit(X_train, y_train)
y_pread3 = svc.predict(X_test)
# decision tree
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(max_depth=5, min_samples_split=10)
tree.fit(X_train, y_train)
y_pread4 = tree.predict(X_test)
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(
    n_estimators=100,
    max_depth=5
)
rf.fit(X_train, y_train)
y_pread5 = rf.predict(X_test)
#gbm
from sklearn.ensemble import GradientBoostingClassifier
gbr = GradientBoostingClassifier(n_estimators=100, max_depth=3)
gbr.fit(X_train, y_train)
y_pread6 = gbr.predict(X_test)
```

```python
from sklearn.metrics import (accuracy_score, precision_score, recall_score, 
                            f1_score, roc_auc_score, confusion_matrix, 
                            classification_report)

# Assuming you have y_test and all your predictions:
# y_pred1, y_pred2, y_pred3, y_pred4, y_pred5, y_pred6

models = {
    'KNN': y_pread1,
    'Logistic Regression': y_pread2,
    'SVM': y_pread3,
    'Decision Tree': y_pread4,
    'Random Forest': y_pread5,
    'Gradient Boosting': y_pread6
}

print("=" * 100)
print("EVALUATION METRICS FOR ALL MODELS")
print("=" * 100)

for model_name, y_pred in models.items():
    print(f"\n{model_name.upper()}")
    print("-" * 100)
    
    # 1. ACCURACY - Overall correctness
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Accuracy:    {accuracy:.4f}  (correct predictions / total predictions)")
    
    # 2. PRECISION - False positives matter
    precision = precision_score(y_test, y_pred)
    print(f"Precision:   {precision:.4f}  (true positives / all positive predictions)")
    
    # 3. RECALL - False negatives matter
    recall = recall_score(y_test, y_pred)
    print(f"Recall:      {recall:.4f}  (true positives / all actual positives)")
    
    # 4. F1-SCORE - Balance between precision and recall
    f1 = f1_score(y_test, y_pred)
    print(f"F1-Score:    {f1:.4f}  (harmonic mean of precision and recall)")
    
    # 5. CONFUSION MATRIX - Detailed breakdown
    cm = confusion_matrix(y_test, y_pred)
    print(f"Confusion Matrix:")
    print(f"  True Negatives:  {cm[0,0]}")
    print(f"  False Positives: {cm[0,1]}")
    print(f"  False Negatives: {cm[1,0]}")
    print(f"  True Positives:  {cm[1,1]}")
    
    # 6. CLASSIFICATION REPORT - Complete summary
    print(f"\nClassification Report:")
    print(classification_report(y_test, y_pred))
```
